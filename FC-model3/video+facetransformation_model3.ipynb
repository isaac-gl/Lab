```python
#xml for face detection and transformation
CASCADE_PATH = "faces.xml"
i = 0
# converting videos into frame in this method
def extractFrames1(pathIn, pathOut, videocount,splt_file,splt_ques):
    print(pathIn)
    cap = cv2.VideoCapture(pathIn)
    count = videocount
    print(splt_file)
    print(splt_ques)
    aucode =np.where((df_expr.ID== str(splt_file))&(df_expr.Con==int(splt_ques)))
    nc = 0
    while (cap.isOpened()):

        # Capture frame-by-frame
        ret, frame = cap.read()
        n = nc//30
        
        if nc%30 == 0:
            flnm = df_expr['AU_Expr'][aucode[0][n]]
        if ret == True:
            
            #print('Read %d frame: ' % count, ret)
            #flname = flnm
            #print(flnm)
            if flnm != 'fake':
                #cv2.imwrite(os.path.join(pathOut, "{:d}frame{:d}.jpg".format(videocount,count)), frame)  # save frame as JPEG file
                generate(frame,True,count,flnm,pathOut,videocount)
                a =0
            count += 1
            nc = nc+1
        else:
            break
        
    return count
            
#generating jpg files by detecting faces from image samples and creating train and test samples          
def generate(image_path, show_result,count,flname,pathOut,videocount):
    img = image_path
    i = count
    face_cascade = cv2.CascadeClassifier(CASCADE_PATH)
    if (img is None):
        #print("Can't open image file")
        return 0
    faces = face_cascade.detectMultiScale(img, 1.1, 3, minSize=(100, 100))
    if (faces is None):
        #print('Failed to detect face')
        return 0

    facecnt = len(faces)
    #print("Detected faces: %d" % facecnt)

    height, width = img.shape[:2]

    for (x, y, w, h) in faces:
        r = max(w, h) / 2
        centerx = x + w / 2
        centery = y + h / 2
        nx = int(centerx - r)
        ny = int(centery - r)
        nr = int(r * 2)

        faceimg = img[ny:ny+nr, nx:nx+nr]
        lastimg = cv2.resize(faceimg, (32, 32))
        i += 1
        #cv2.imwrite("C:/Users/Vijay/Isaac/Issac Vedio/data/%d_%s.jpg.jpg" % i %flname, lastimg)
        cv2.imwrite(os.path.join(pathOut, "{:d}_{:s}.jpg".format(count,flname)), lastimg)
```
