```python
def tm_res(df1,videoID,Con):
    lst = []
    new_df1 = df1.loc[df1['ID_val'] == videoID].loc[df1['Con'] == Con]
    p = ['AU1','AU2','AU4','AU5','AU6','AU7','AU9','AU10','AU11','AU12','AU14','AU15','AU16','AU17','AU22','AU23',
         'AU24','AU25','AU26','AU27']

    for i in range(new_df1.shape[0]):
        u_lst = []
        #print(i)
        for j in range(len(p)):
            #print(j)
            if float(new_df1.iloc[i][str(p[j])])>0:
                u_lst.append(new_df1.columns[j+3])
        if len(u_lst) == 0:
            u_lst.append('fake')
        lst.append(u_lst)
    new_df1['AU_Val'] = np.array(lst)
    return new_df1

df_new = []
for i in range(len(dirlist)):
    print(dirlist[i])
    final_data1 = tm_res(df1,dirlist[i],1)
    final_data2 = tm_res(df1,dirlist[i],2)
    final_data3 = tm_res(df1,dirlist[i],3)
    final_data4 = tm_res(df1,dirlist[i],4)
    df_final = final_data1.append([final_data2, final_data3, final_data4], ignore_index=True)
    df_new.append(df_final)

df_new1 = pd.concat(df_new, ignore_index=True)
```

```python
#xml for face detection and transformation
CASCADE_PATH = "faces.xml"
i = 0
# converting videos into frame in this method
def extractFrames1(pathIn, pathOut, videocount,splt_file,splt_ques):
    print(pathIn)
    cap = cv2.VideoCapture(pathIn)
    count = videocount
    print(splt_file)
    print(splt_ques)
    aucode =np.where((df_expr.ID_val== str(splt_file))&(df_expr.Con==int(splt_ques)))
    nc = 0
    while (cap.isOpened()):

        # Capture frame-by-frame
        ret, frame = cap.read()
        n = nc//30
        
        if nc%30 == 0:
            flnm = df_expr['AU_Expr'][aucode[0][n]]
        if ret == True:
            
            #print('Read %d frame: ' % count, ret)
            #flname = flnm
            #print(flnm)
            if flnm != 'fake':
                #cv2.imwrite(os.path.join(pathOut, "{:d}frame{:d}.jpg".format(videocount,count)), frame)  # save frame as JPEG file
                generate(frame,True,count,flnm,pathOut,videocount)
                a =0
            count += 1
            nc = nc+1
        else:
            break
        
    return count
            
#generating jpg files by detecting faces from image samples and creating train and test samples          
def generate(image_path, show_result,count,flname,pathOut,videocount):
    img = image_path
    i = count
    face_cascade = cv2.CascadeClassifier(CASCADE_PATH)
    if (img is None):
        #print("Can't open image file")
        return 0
    faces = face_cascade.detectMultiScale(img, 1.1, 3, minSize=(100, 100))
    if (faces is None):
        #print('Failed to detect face')
        return 0

    facecnt = len(faces)
    #print("Detected faces: %d" % facecnt)

    height, width = img.shape[:2]

    for (x, y, w, h) in faces:
        r = max(w, h) / 2
        centerx = x + w / 2
        centery = y + h / 2
        nx = int(centerx - r)
        ny = int(centery - r)
        nr = int(r * 2)

        faceimg = img[ny:ny+nr, nx:nx+nr]
        lastimg = cv2.resize(faceimg, (32, 32))
        i += 1
        #cv2.imwrite("C:/Users/Vijay/Isaac/Issac Vedio/data/%d_%s.jpg.jpg" % i %flname, lastimg)
        cv2.imwrite(os.path.join(pathOut, "{:d}_{:s}.jpg".format(count,flname)), lastimg)
```
